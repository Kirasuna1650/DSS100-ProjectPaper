````python

import pandas as pd
import numpy as np
import re
from IPython.display import display

# Set the exact CSV file you want to load (forward slashes work on Windows in Python)
CSV_PATH = r'C:/Users/ajmcs/OneDrive/VSCODE School Projects/25-26Q1/Jupyter/csv files/StudentsPerformance (1).csv'

def find_column(df, patterns):
    cols = df.columns.astype(str)
    for pat in patterns:
        regex = re.compile(pat, flags=re.I)
        for c in cols:
            if regex.search(c):
                return c
    return None

try:
    df = pd.read_csv(CSV_PATH)
except FileNotFoundError:
    raise RuntimeError(f'CSV not found: {CSV_PATH}\nPlease verify the path or run Cell 1 to list available CSV files.')

print('Loaded dataframe with shape:', df.shape)
display(df.head())

# try to detect common columns
gender_col = find_column(df, [r'\bgender\b', r'\bsex\b', r'male|female'])
race_col = find_column(df, [r'race', r'ethnic', r'ethnicity'])
math_col = find_column(df, [r'math', r'math score', r'math_score'])
reading_col = find_column(df, [r'read', r'reading', r'reading score', r'reading_score'])
writing_col = find_column(df, [r'write', r'writing', r'writing score', r'writing_score'])

print('Detected columns:')
print(' gender:', gender_col)
print(' race/ethnicity:', race_col)
print(' math:', math_col)
print(' reading:', reading_col)
print(' writing:', writing_col)
````

````python
# Cell 3: Data preprocessing
# - Check for missing values and basic imputation
# - Remove duplicates

# summary of missing values
missing = df.isna().sum()
print('Missing values per column:', missing[missing>0])

# drop exact duplicate rows
before = len(df)
df = df.drop_duplicates().reset_index(drop=True)
print(f'Dropped {before - len(df)} duplicate rows')

# Coerce numeric columns where possible and identify numeric score columns
for c in df.columns:
    # try convert to numeric when majority look numeric
    try:
        df[c] = pd.to_numeric(df[c], errors='ignore')
    except Exception:
        pass

# Identify numeric score columns if detection above failed
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
possible_scores = [c for c in numeric_cols if re.search(r'math|read|write|score', c, flags=re.I)]
print('Numeric columns detected:', numeric_cols)
print('Possible score columns:', possible_scores)

# Impute numeric columns with median and categorical with mode
for c in df.columns:
    if df[c].dtype.kind in 'biufc':  # numeric
        if df[c].isna().any():
            med = df[c].median()
            df[c] = df[c].fillna(med)
            print(f'Imputed numeric column {c} with median={med}')
    else:
        if df[c].isna().any():
            mode = df[c].mode(dropna=True)
            if not mode.empty:
                df[c] = df[c].fillna(mode.iloc[0])
                print(f'Imputed categorical column {c} with mode={mode.iloc[0]}')

print('Preprocessing complete. Current shape:', df.shape)
````

````python
# Cell 4: Visualizations (pie, bar, scatter, line)
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Prepare columns (use detected ones if available)
gcol = gender_col or 'gender'
rcol = race_col or 'race/ethnicity'
mcol = math_col or (possible_scores[0] if possible_scores else None)
recol = reading_col or (possible_scores[1] if len(possible_scores)>1 else None)
wcol = writing_col or (possible_scores[2] if len(possible_scores)>2 else None)

# Pie chart: distribution by gender (if available)
if gcol and gcol in df.columns:
    fig, ax = plt.subplots(figsize=(6,6))
    df[gcol].value_counts().plot.pie(autopct='%1.1f%%', ax=ax)
    ax.set_ylabel('')
    ax.set_title('Distribution by ' + gcol)
    plt.show()
else:
    print('Gender column not found; skipping pie chart')

# Bar graph: average scores by gender or race
group_col = gcol if (gcol in df.columns) else (rcol if rcol in df.columns else None)
score_cols = [c for c in [mcol,recol,wcol] if c and c in df.columns]
if group_col and score_cols:
    means = df.groupby(group_col)[score_cols].mean()
    means.plot.bar(figsize=(8,5))
    plt.title('Average scores by ' + group_col)
    plt.ylabel('Average score')
    plt.show()
else:
    print('Insufficient columns for grouped bar chart (need group_col and score columns).')

# Scatter: math vs reading
if mcol and recol and mcol in df.columns and recol in df.columns:
    plt.figure(figsize=(7,5))
    sns.scatterplot(data=df, x=mcol, y=recol, hue=group_col if group_col in df.columns else None, alpha=0.7)
    plt.title(f'{mcol} vs {recol}')
    plt.show()
else:
    print('Skipping scatter plot: math or reading column missing')

# Line graph: trend of mean scores across race/ethnicity groups (if available)
if rcol and rcol in df.columns and score_cols:
    means_race = df.groupby(rcol)[score_cols].mean().T
    means_race.plot(figsize=(8,4))
    plt.title('Score trends across ' + rcol)
    plt.xlabel('Subject')
    plt.ylabel('Average score')
    plt.show()
else:
    print('Skipping line graph: race column or score columns missing')
````

````python
# Cell 5: Exploratory Data Analysis (central tendency & dispersion, boxplot, correlation heatmap)
from scipy import stats

num_cols = [c for c in df.columns if df[c].dtype.kind in 'biufc']
print('Numeric columns used for EDA:', num_cols)

# central tendencies
eda = pd.DataFrame(index=['mean','median','mode','std','var'])
for c in num_cols:
    mode = df[c].mode(dropna=True)
    eda[c] = [df[c].mean(), df[c].median(), (mode.iloc[0] if not mode.empty else np.nan), df[c].std(), df[c].var()]

display(eda)

# IQR and boxplot
iqr = {c: stats.iqr(df[c].dropna()) for c in num_cols}
print('IQR per numeric column:', iqr)

plt.figure(figsize=(8,5))
sns.boxplot(data=df[num_cols])
plt.title('Boxplots of numeric columns')
plt.show()

# Correlation heatmap
corr = df[num_cols].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation matrix')
plt.show()

# Guiding questions (printed summary)
print('Guiding answers:')
if num_cols:
    means = df[num_cols].mean()
    print(' Subject with highest average score:', means.idxmax(), '->', means.max())
    print(' Subject with most variability (std):', df[num_cols].std().idxmax(), '->', df[num_cols].std().max())
    if 'math' in ' '.join(num_cols).lower() and 'reading' in ' '.join(num_cols).lower():
        print('Correlation between math and reading:', corr.get(mcol, {}).get(recol) if mcol and recol else corr)
else:
    print('No numeric columns found for EDA')
````

````python
# Cell 6: Normality checks (skewness & kurtosis, histograms, Q-Q plots)
from scipy.stats import skew, kurtosis, probplot

for c in num_cols:
    s = skew(df[c].dropna())
    k = kurtosis(df[c].dropna())
    print(f'{c}: skewness={s:.3f}, kurtosis={k:.3f}')
    plt.figure(figsize=(6,3))
    plt.subplot(1,2,1)
    sns.histplot(df[c].dropna(), kde=True)
    plt.title(f'{c} distribution')
    plt.subplot(1,2,2)
    probplot(df[c].dropna(), dist='norm', plot=plt)
    plt.title('Q-Q plot')
    plt.tight_layout()
    plt.show()

# Quick interpretive guidance printed:
print('Interpretation guidance:')
print('- Skewness near 0: approx symmetric. Positive: right skew. Negative: left skew.')
print('- Kurtosis > 0: heavier tails (outliers more likely).')
````

````python
# Cell 7: Clustering (K-Means) on numeric score columns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X = df[num_cols].dropna()
if X.shape[1] < 1 or X.shape[0] < 2:
    print('Not enough numeric columns or rows for clustering')
else:
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)
    best_k = None
    best_score = -1
    scores = {}
    for k in range(2, min(7, len(X))):
        km = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = km.fit_predict(Xs)
        try:
            s = silhouette_score(Xs, labels)
        except Exception:
            s = -1
        scores[k] = s
        if s > best_score:
            best_score = s
            best_k = k
    print('Silhouette scores by k:', scores)
    print('Best k by silhouette:', best_k, 'score=', best_score)

    # Fit final KMeans and attach labels to dataframe copy for plotting
    km = KMeans(n_clusters=best_k, random_state=42, n_init=10)
    labs = km.fit_predict(Xs)
    df_clustered = X.copy()
    df_clustered['cluster'] = labs
    # plot first two numeric dims
    plt.figure(figsize=(7,5))
    sns.scatterplot(x=Xs[:,0], y=Xs[:,1], hue=labs, palette='tab10', alpha=0.7)
    plt.title(f'KMeans clusters (k={best_k}) on first two scaled numeric dims')
    plt.show()

    # cluster summary
    print('Cluster centers (scaled):')
    print(km.cluster_centers_)
    print('Cluster sizes:')
    print(df_clustered['cluster'].value_counts())
````
